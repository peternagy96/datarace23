```python
# TARGET_EVENT = K means the loan was defaulter
# TARGET_EVENT = E means the loan was prepaid
# TARGET_EVENT = - means the loan is still active
# get only data from borrowers who have defaulted or prepaid

df_train = df[df.BORROWER_ID.isin(df[df.TARGET_EVENT != '-'].BORROWER_ID.unique())]
```

```python
df_train.TARGET_EVENT_DAY = pd.to_datetime(df_train.TARGET_EVENT_DAY, origin='julian', unit='D')
```

```python
df_train.BORROWER_ID.nunique()
```

```python
# calculate the percentage of the target variable being defaulted per borrower
df_targets = (df_train.groupby('BORROWER_ID').TARGET_EVENT.value_counts() / df_train.groupby('BORROWER_ID').TARGET_EVENT.count()).reset_index().rename(columns={0: 'label'}).sort_values('label', ascending=False)
df_targets = df_targets[df_targets.TARGET_EVENT == 'K'][['BORROWER_ID', 'label']]
```

```python
df_train = df_train.merge(df_targets, on='BORROWER_ID', how='left')
df_train.label = df_train.label.fillna(0)
```

```python
df_train.label.value_counts().sort_index()
```

```python
df_train.columns
```

```python
df_train = df_train[[
        'BORROWER_ID', 'CONTRACT_BANK_ID',
       'CONTRACT_CREDIT_INTERMEDIARY', 'CONTRACT_CREDIT_LOSS',
       'CONTRACT_CURRENCY', 'CONTRACT_DATE_OF_LOAN_AGREEMENT',
       'CONTRACT_INCOME', 'CONTRACT_INSTALMENT_AMOUNT',
       'CONTRACT_INSTALMENT_AMOUNT_2', 'CONTRACT_INTEREST_PERIOD',
       'CONTRACT_INTEREST_RATE', 'CONTRACT_LGD', 'CONTRACT_LOAN_AMOUNT',
       'CONTRACT_LOAN_CONTRACT_TYPE', 'CONTRACT_LOAN_TO_VALUE_RATIO',
       'CONTRACT_LOAN_TYPE', 'CONTRACT_MARKET_VALUE',
       'CONTRACT_MORTGAGE_LENDING_VALUE', 'CONTRACT_MORTGAGE_TYPE',
       'CONTRACT_REFINANCED', 'CONTRACT_RISK_WEIGHTED_ASSETS',
         'BORROWER_BIRTH_YEAR',
       'BORROWER_CITIZENSHIP', 'BORROWER_COUNTRY', 'BORROWER_COUNTY',
       'BORROWER_TYPE_OF_CUSTOMER',
       'TARGET_EVENT', 'TARGET_EVENT_DAY', 'label'
]]
```

```python
df_train.info()
```

```python
categorical_columns = ['CONTRACT_BANK_ID', 'CONTRACT_LOAN_TYPE', 'BORROWER_TYPE_OF_CUSTOMER']

# encode categorical columns
df_train = pd.get_dummies(df_train, columns=categorical_columns, drop_first=True)
```

```python
df_train
```

```python
# train an XGBoost model
import xgboost as xgb

# split the data into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_train.drop(['BORROWER_ID', 'TARGET_EVENT', 'TARGET_EVENT_DAY', 'label'], axis=1), df_train.label, test_size=0.1, random_state=42)


# define the model
model = xgb.XGBRegressor(objective='binary:logistic', n_estimators=1000, seed=42, max_depth=5, learning_rate=0.1, gamma=0.1, reg_lambda=0.1, reg_alpha=0.1, subsample=0.8, colsample_bytree=0.8)

# fit the model
model.fit(X_train, y_train, eval_metric='auc', eval_set=[(X_train, y_train), (X_test, y_test)], early_stopping_rounds=50)
```

```python
# plot feature importance
from xgboost import plot_importance
import matplotlib.pyplot as plt

plot_importance(model, max_num_features=20)
```

```python
# plot shap values on test data

import shap

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)

shap.summary_plot(shap_values, X_test, max_display=20)
```
