## Non trivial variables
- CONTRACT_DEPT_SERVICE_TO_INCOME - "The higher the ratio, the larger the portion of the borrower's income that is dedicated to servicing the debt. A high ratio indicates that a significant portion of the borrower's income is going towards repaying debts, which could potentially increase the risk of default."
- CONTRACT_INSTALMENT_AMOUNT - törlesztőrészlet "A high installment amount relative to the borrower's income may increase the risk of default if the borrower faces financial difficulties. On the other hand, a reasonable and manageable installment amount enhances the likelihood that the borrower can consistently meet their repayment obligations."
- CONTRACT_LGD - loss given default "a measure of the potential loss a lender might face in the event of a borrower defaulting on a loan, higher LGD implies a greater potential loss for the lender in the event of a default."
- CONTRACT_CREDIT_LOSS - Credit loss represents the financial loss a lender may incur if a borrower fails to repay a loan. It is a critical metric in credit risk management and is typically calculated based on the probability of default (PD), loss given default (LGD), and exposure at default (EAD). The formula for credit loss is often expressed as: Credit Loss = PD×LGD×EAD
- CONTRACT_LOAN_TO_VALUE_RATIO - LTV "is a key metric in real estate and mortgage lending that assesses the relationship between the loan amount and the appraised value of the property, providing insights into the risk associated with the loan."
- CONTRACT_RISK_WEIGHTED_ASSETS - "risk-weighted assets helps financial institutions determine the amount of capital they need to hold to cover the risks associated with their lending and investment activities."

## Other concepts
- Probability of Default (PD): This is an estimate of the likelihood that a borrower will default on a loan within a specific time frame. It is a key input into the credit loss calculation.
- Loss Given Default (LGD): As discussed in a previous response, LGD represents the proportion of the exposure at default that the lender expects to lose if a borrower defaults.
- Exposure at Default (EAD): This is the total amount at risk when a borrower defaults. It is the outstanding loan amount at the time of default.
